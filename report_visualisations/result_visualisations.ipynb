```python
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Fixed intervals

#### 120s

```python
interval = 120
var = f'24_{interval}'

# Load data
with open(f"../data/LDZO/06_monte_carlo/{var}/total_counts_aggregated.pkl", "rb") as f:
    ldzo_120 = pickle.load(f)
with open(f"../data/LHCC/06_monte_carlo/{var}/total_counts_aggregated.pkl", "rb") as f:
    lhcc_120 = pickle.load(f)
with open(f"../data/LMMM/06_monte_carlo/{var}/total_counts_aggregated.pkl", "rb") as f:
    lmmm_120 = pickle.load(f)
with open(f"../data/LSAZ/06_monte_carlo/{var}/total_counts_aggregated.pkl", "rb") as f:
    lsaz_120 = pickle.load(f)

# combine data
data = [ldzo_120, lhcc_120, lmmm_120, lsaz_120]

# create and show plot
plt.figure(figsize=(15, 8))
sns.boxplot(data=data, showfliers=False, palette="pastel", notch= True, showmeans=True)
plt.ylabel('LOS occurence count')
plt.xticks(range(4), ['LDZO', 'LHCC', 'LMMM', 'LSAZ'])
plt.title(f"Distribution of occurence counts for {interval} second injection intervals")
plt.grid(True)
plt.show()
```

#### Overview injection intervals

```python
import pandas as pd

# Combine data
# create a list for each airspace
ldzo_all = [np.mean(ldzo_120), np.mean(ldzo_100), np.mean(ldzo_80), np.mean(ldzo_60), np.mean(ldzo_40), np.mean(ldzo_20)]
lhcc_all = [np.mean(lhcc_120), np.mean(lhcc_100), np.mean(lhcc_80), np.mean(lhcc_60), np.mean(lhcc_40), np.mean(lhcc_20)]
lmmm_all = [np.mean(lmmm_120), np.mean(lmmm_100), np.mean(lmmm_80), np.mean(lmmm_60), np.mean(lmmm_40), np.mean(lmmm_20)]
lsaz_all = [np.mean(lsaz_120), np.mean(lsaz_100), np.mean(lsaz_80), np.mean(lsaz_60), np.mean(lsaz_40), np.mean(lsaz_20)]

# combine to a dictionary
data = {
    'LDZO': ldzo_all,
    'LHCC': lhcc_all,
    'LMMM': lmmm_all,
    'LSAZ': lsaz_all
}

# Turn dictionary into a DataFrame
x_labels = ['120', '100', '80', '60', '40', '20']
df = pd.DataFrame(data, index=x_labels)

# Plot data
plt.figure(figsize=(15, 8))
sns.lineplot(data=df, dashes=False, marker='o')
plt.xlabel('Injection interval [s]')
plt.ylabel('Mean LOS occurence count')
plt.title('Mean LOS occurence count for different injection intervals')
plt.grid(True)
plt.show()
```

#### Overview number of crossing AC

```python
import pandas as pd

# Combine data
# create a list for each airspace
ldzo_all = [np.mean(ldzo_120), np.mean(ldzo_100), np.mean(ldzo_80), np.mean(ldzo_60), np.mean(ldzo_40), np.mean(ldzo_20)]
lhcc_all = [np.mean(lhcc_120), np.mean(lhcc_100), np.mean(lhcc_80), np.mean(lhcc_60), np.mean(lhcc_40), np.mean(lhcc_20)]
lmmm_all = [np.mean(lmmm_120), np.mean(lmmm_100), np.mean(lmmm_80), np.mean(lmmm_60), np.mean(lmmm_40), np.mean(lmmm_20)]
lsaz_all = [np.mean(lsaz_120), np.mean(lsaz_100), np.mean(lsaz_80), np.mean(lsaz_60), np.mean(lsaz_40), np.mean(lsaz_20)]

# combine to a dictionary
data = {
    'LDZO': ldzo_all,
    'LHCC': lhcc_all,
    'LMMM': lmmm_all,
    'LSAZ': lsaz_all
}

# Turn dictionary into a DataFrame
x_labels = [720, 864, 1080, 1440, 2160, 4320]
df = pd.DataFrame(data, index=x_labels)
df = df.reset_index().rename(columns={'index': 'number_of_aircraft'})

# Plot data
plt.figure(figsize=(15, 8))
plt.xlabel('Number of crossing aircraft within 24 hours')
plt.ylabel('Mean LOS occurence count')
plt.title('Mean LOS occurence count for different number of aircraft crossing the airspace within 24 hours')
# Add legend and line for each element
for column in df.columns[1:]:
    sns.lineplot(data=df, x='number_of_aircraft', y=column, label=column, marker='o')
plt.grid(True)
plt.show()
```
