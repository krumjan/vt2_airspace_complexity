```python
import glob
from traffic.core import Traffic
import pandas as pd
from tqdm.auto import tqdm

i=5

folder = f'/cluster/home/krum/github/VT2_airspace_complexity/data/LSAZ/01_raw/2019_{i}'

files = glob.glob(f"{folder}/*.parquet", recursive=True)

traflist = []

for file in tqdm(files):
    print(file)
    traf = Traffic.from_file(file)
    traflist.append(traf.data)

alldata = Traffic(pd.concat(traflist, ignore_index=True))

path_combined = '/cluster/home/krum/github/VT2_airspace_complexity/data/LSAZ/02_combined'
year = 2019
month = i

alldata.to_parquet(
    f"{path_combined}/combined_{year}_{month}.parquet"
)
```
